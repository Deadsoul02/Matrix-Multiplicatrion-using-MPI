```python
import numpy as np
import matplotlib.pyplot as plt

# createdata
x = np.array([300, 600, 900, 1200, 1500, 1800, 2100, 2400, 2700, 3000])  #Size of Matrix
y1 = np.array([0.021502971649169922,0.13722586631774902,0.5105879306793213,2.270228862762451,3.580112934112549,6.7340312004089355,16.87362790107727,20.609497785568237,35.689615964889526,78.16089200973511]) #time taken by Serial Multiplication
y2 = np.array([0.004302978515625,0.046568870544433594, 0.31837987899780273,0.9947388172149658,1.8092429637908936, 3.1415538787841797,6.604243040084839,9.389286041259766,14.434798240661621,24.35834503173828]) #time taken by Parallel Multiplication


# set figure size
plt.figure(figsize=(10, 6))

# plot the data
plt.plot(x, y1, label='Serial Multiplication')
plt.plot(x, y2, label='Parallel Multiplication')

# add labels and legend
plt.xlabel('Matrix Size')
plt.ylabel('Time-Taken')
plt.title('Serial vs Parallel Multiplication using MPI')
plt.legend(fontsize='large')

# display the graph
plt.show()
```

```python
import numpy as np
from tabulate import tabulate

# createdata
x = np.array([300, 600, 900, 1200, 1500, 1800, 2100, 2400, 2700, 3000])  #Size of Matrix
y1 = np.array([0.021502971649169922,0.13722586631774902,0.5105879306793213,2.270228862762451,3.580112934112549,6.7340312004089355,16.87362790107727,20.609497785568237,35.689615964889526,78.16089200973511]) #time taken by Serial Multiplication
y2 = np.array([0.004302978515625,0.046568870544433594, 0.31837987899780273,0.9947388172149658,1.8092429637908936, 3.1415538787841797,6.604243040084839,9.389286041259766,14.434798240661621,24.35834503173828]) #time taken by Parallel Multiplication





# create table
table = []
for i in range(len(x)):
    table.append([x[i], y1[i], y2[i], y1[i]/y2[i]])

# print table
headers = ["Matrix Size", "Time Taken by Serial Multiplication", "Time Taken by Parallel Multiplication", "Speed Up"]
print(tabulate(table, headers=headers ))
```
